### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_file_path = 'resources/city_data.csv'
ride_file_path = 'resources/ride_data.csv'

# Read the City and Ride Data
city_data = pd.read_csv(city_file_path)
ride_data = pd.read_csv(ride_file_path)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data = pd.merge(ride_data, city_data, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
rider_total_type = pyber_data.groupby('type')['ride_id'].count()
rider_total = rider_total_type.sum()
rider_total_type
# print(f'Total number of rides: {rider_total}')
# Total number of rides: 2375
```

```python
# 2. Get the total drivers for each city type
driver_total_type = city_data.groupby('type')['driver_count'].sum()
driver_total = driver_total_type.sum()
driver_total_type
# print(f'Total number of drivers: {driver_total}')
# Total number of drivers: 2973
```

```python
#  3. Get the total amount of fares for each city type
fare_total_type = pyber_data.groupby('type')['fare'].sum()
fare_total = fare_total_type.sum()
fare_total_type
# print(f'Total fares: {fare_total}')
# Total fares: 63538.64
```

```python
#  4. Get the average fare per ride for each city type. 
rider_fare_mean_type = fare_total_type / rider_total_type
rider_fare_mean_type
```

```python
# 5. Get the average fare per driver for each city type. 
driver_fare_mean_type = fare_total_type / driver_total_type
driver_fare_mean_type
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary = pd.DataFrame({
    'Total Rides': rider_total_type,
    'Total Drivers': driver_total_type,
    'Total Fares': fare_total_type.map("${:,.2f}".format),
    'Average Fare per Rider': rider_fare_mean_type.map("${:,.2f}".format),
    'Average Fare per Driver': driver_fare_mean_type.map("${:,.2f}".format),
})
pyber_summary.head(10)
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary.index.name = None
```

```python
#  8. Format the columns.
# this was done on creation of the DataFrame to ensure that I wouldn't forget :) 
pyber_summary.head(10)
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data = pd.merge(ride_data, city_data, how="left", on=["city", "city"])

# Display the DataFrame
pyber_data.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
sum_pyber_type_date = pyber_data.groupby(['type', 'date'])[['fare']].sum()
sum_pyber_type_date
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
sum_pyber_type_date = sum_pyber_type_date.reset_index()
sum_pyber_type_date
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
sum_pyber_type_date = sum_pyber_type_date.pivot(index='date', columns='type', values='fare')
sum_pyber_type_date.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
sum_pyber_data = pd.DataFrame(sum_pyber_type_date)
sum_pyber_data = sum_pyber_data.loc['2019-01-01':'2019-04-29']
sum_pyber_data.head()
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
sum_pyber_data.index = pd.to_datetime(sum_pyber_data.index)
sum_pyber_data
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
sum_pyber_data.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
sum_pyber_data = sum_pyber_data.resample('W').sum()
sum_pyber_data.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

ax = sum_pyber_data.plot(figsize=(11,4))
ax.set_title('Total Fare by City Type',fontsize=20)
ax.set_ylabel('Fare (USD)',fontsize=14)
ax.set_xlabel("Time",fontsize=14)

plt.savefig("analysis/PyBer_fare_summary.png")
```
